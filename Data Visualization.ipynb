## Basic Graphs

To create a basic graph, you will need to import the *pyplot* library within *matplotlib*. You will also need list data types to pass through the *plot()* method in matplotlib which will create the graph for you. First parameter in *plot* is the x axis and second is the y axis. Now, the graph is created but it has to displayed on the screen. We can now use *show* function to display the graph.

```python
import matplotlib.pyplot as plt  # import matlib with plot function and rename as plt

years = [1, 1000, 1500, 1600, 1700, 1750, 1800, 1850, 1900, 1950,    # data being used
         1955, 1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]

population = [200, 400, 450, 500, 682, 791, 1000, 1262, 1650, 2525, 2750, 3010, 3322, 3682, 
              4061, 4440, 4853, 5310, 5735, 6127, 6520, 6930, 7349]

plt.plot(years, population)  # passing the list data types into the plot function from plt which is matplotlib.pyplot
plt.show() # displays the graph
```

## Adding Labels and Color

Labels can be added to your graph using matplotlib. Using the pyplot library within matplotlib, we can create labels. 
If we *import matplotlib.pyplot as plt* just as in the previous code, we can create labels for the x and y axis using *plt.xlabel()* and *plt.ylabel()* respectively and declaring the label as a string within the parenthesises. A title for the graph can also be created using *plt.title()*. Another feature from pyplot is the color of the line on the graph. A parameter can be added to *plt.plot()* after the two data types are taken as parameters. The parameter to be added is *color*. *color* is a method that takes three parameters which determine the tone of the line. They are red, green and blue in that order. To determine the tone, create a fraction with 255 as denominator and any number at the numerator and pass it for each parameter in the *color* method in order to get any color on the color spectrum. If all parameters are at 255/255, the line will be white and if all parameters are 0/255, the line will display black.  The addition of labels and color in the line are demonstrated in the code below.

```python
import matplotlib.pyplot as plt

years = [1950, 1955, 1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]
population = [2.525, 2.750, 3.010, 3.322, 3.682, 4.061, 4.440, 4.853, 5.310, 5.735, 6.127, 6.520, 6.930, 7.349]

plt.xlabel("Population Growth by Year")  # creating label for x axis
plt.ylabel("Population in Billions")  #creating label for y axis

plt.title("Population Growth 1950-2015")  #creating title of the graph

plt.plot(years, population, color=(0/255, 230/255, 0/255))  # third parameter of color being passed
plt.show()
```

## Axis and Grid

The axis of a graph can be edited using the *axis()* which comes from *pyplot*. Axis takes a list data type as the argument. Within the list, place your values as such [xmin, xmax, ymin, ymax] and this will scale your graph. Additionally, a grid can be added to the graph by running *pyplot.grid(True)*.

```python
import matplotlib.pyplot as plt

years = [1950, 1955, 1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]
population = [2.525, 2.750, 3.010, 3.322, 3.682, 4.061, 4.440, 4.853, 5.310, 5.735, 6.127, 6.520, 6.930, 7.349]

plt.xlabel("Population Growth by Year")  
plt.ylabel("Population in Billions")
plt.title("Population Growth 1950-2015")

plt.axis([1945, 2020, 2, 8]) # setting the scale of the axis
plt.grid(True)              # adding grid to graph

plt.plot(years, population, color=(0/255, 230/255, 0/255))
plt.show()
```

## Multiple Lines and Line Style

Multiple lines can be displayed on a graph. This can be done by using the *plot()* method more than once. Another cool thing to do is change the style of the line. Instead of having a straight line, a dashed line can also be created by passing '--' in the *color* method. Other commands can also be passed such as '-' | '--' | '-.' | ':' | '.' to style the line.

```python
import matplotlib.pyplot as plt

years = [1950, 1955, 1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]
population = [2.525, 2.750, 3.010, 3.322, 3.682, 4.061, 4.440, 4.853, 5.310, 5.735, 6.127, 6.520, 6.930, 7.349]
deaths = [.9, .998, 1.3, 1.5, 1.898, 1.993, 2.04, 2.13, 2.25, 2.57, 2.67, 2.856, 2.993, 3.08] # new data for new line

plt.xlabel("Population Growth and Death by Year") 
plt.ylabel("Population Growth and Death in Billions")  

plt.title("World Population 1950-2015") 

plt.plot(years, population, color=(0/255, 230/255, 0/255)) 
plt.plot(years, deaths, ':' , color=(255/255, 80/255, 170/255)) # second line created on graph as dashed
plt.show()
```

## Other Line Configurations

We can further edit the graph using the method *setp()* from the pyplot library. This lets you further set the properties of a line. Firstly, initialize the *plot()* method with its parameters to a variable. Then, pass the variable into the *setp()* method. This method will take more arguments that would edit the line on the graph. Such arguments to pass would be *color, linewidth, marker, markersize, linestyle* etc. I will demonstrate how to use these arguments stated in the code below. If you would like to check out more configurations for the line, go to this link --https://matplotlib.org/tutorials/introductory/pyplot.html#sphx-glr-tutorials-introductory-pyplot-py. Once here, scroll to Controlling line properties section. 

```python
import matplotlib.pyplot as plt

years = [1950, 1955, 1960, 1965, 1970, 1975, 1980, 1985, 1990, 1995, 2000, 2005, 2010, 2015]
population = [2.525, 2.750, 3.010, 3.322, 3.682, 4.061, 4.440, 4.853, 5.310, 5.735, 6.127, 6.520, 6.930, 7.349]
deaths = [.9, .998, 1.3, 1.5, 1.898, 1.993, 2.04, 2.13, 2.25, 2.57, 2.67, 2.856, 2.993, 3.08]

plt.xlabel("Population Growth and Death by Year") 
plt.ylabel("Population Growth and Death in Billions")  

plt.title("World Population 1950-2015") 

plt.grid(True)

lines = plt.plot(years, population, years, deaths) # can take both lines as parameters

# both lines will have these configurations
plt.setp(lines, color=(.1, .5, .1), linewidth = 1.0, marker = 'o', markersize = 10.0, linestyle = '-') 
plt.show()
```

## Intro to Pie Charts

The *pie()* method is used to create pie charts which is inherited from the pyplot library. Firstly, pass the actual data into the first parameter in order to build the chart and then pass all the labels for each piece of data and set it equal to the *labels* parameter. The *autopct* argument is set equal to '%1.1f%%'. This serves as a place holder for the percentage value that will be calculated for the function. The number value placed after the decimal and before the letter f determine the amount of decimal places that will be displayed for each piece of data.

```python
import matplotlib.pyplot as plt

labels = ['Python', 'C++', 'Java', 'Ruby', 'Perl', 'C#']
people = ['68', '18', '50', '17', '45', '31']

# first pass in data and then pass labels equal to labels parameter
plt.pie(people, labels=labels, autopct='%0.1f%%')

plt.axis('equal') # if chart window were to stretch such as in Pycharm Terminal, chart will not stretch with window

plt.show()
```

One can also remove one slice of the pie chart by setting the *explode* argument in the *pie()* method to a list for each piece of data. Each value placed in the list will correspond to the list passed as the first argument. If the value in the list is set to zero then the slice will stay in the chart. Any value higher than zero will remove the slice further and further from the chart.

```python
import matplotlib.pyplot as plt

labels = ['Python', 'C++', 'Java', 'Ruby', 'Perl', 'C#']
people = ['68', '18', '50', '17', '45', '31']


plt.pie(people, labels=labels, autopct='%0.1f%%', explode=[0,0,0,.1,0,0])
# here, 'Ruby' is removed from the slice because its value is greater than zero

plt.show()
```

Pandas allows you to create a list as a dictionary definition. First you import the pandas library by typing *import pandas* in your python code. Afterwards, create a dictionary with each key name having a list as its definition. Now call the *pandas* library and using the *DataFrame()* function from the pandas library, pass in as parameters your dictionary and then a second paramatar set equal to *columns*. The key name in the dictionary is what is the column is set to. After both parameters are passed, the method will organize the dictionary into a table with all the key names and list definitions. In order to depict what the function does, we will print it out. Here is an example in code to better explain this.

## Using Pandas for Pie Charts

```python
import pandas
import matplotlib.pyplot as plt

dictionar = {'name': ['Osher', 'Layla', 'Fred', 'Talos', 'Seabass'],
    'b-day': ['04/22', '12/09', '09/13', '02/29', '10/10'],
    'cell': ['8095556789', '1234567892', '1223547782', '8752947592','7764428742'],
    'days missed' : [4, 6, 7, 7, 2],
    'days off' : [10, 10, 10, 11, 12]} 

data = pandas.DataFrame(dictionar, columns=['name', 'b-day', 'cell', 'days missed', 'days off'])  

data['total days off'] = data['days missed'] + data['days off']

#creating list variable for each color of the pie
color = [(.1, .6, .75), (.9,.6,.7), (1.0, .4, 0.2), (0.3, 1.0, 0.85), (1.0, .08, 1.0)] 

# creating pie chart by passing in the data variable and indexing the data that will printed as well as adding labels etc
plt.pie(data['total days off'], labels=data['name'], colors=color, autopct='%1.1f%%')#

plt.show()
```

## Bar Charts

In order to use bar charts, the numpy python library must be imported. This is done so that we can call the *arange* method which will create a type of array with even spaced values which will allow the bar chart to be created from the data. Then we pass it into the *bar()* method from the pyplot class along with the data needed to create bar chart.

```python
import matplotlib.pyplot as plt
import numpy         

scores = [554, 654, 606]

#creating variable for numpy method with amount of data being passed
index = numpy.arange(3)

# passing in numpy variable so it can be used to create the bars on the chart
plt.bar(index, scores, color=(.5, .5, .7))
plt.grid(True)

plt.show()
```

To represent more than one piece of data, use the *bar()* method again. The issue with doing this though is that the bars will print on top of eachother. In order to make it look better while still keeping the bars on top of eachother is by using the third parameter in the *bar()* method which will manipulate the bar width.  You can also just set the value need to *width* within the *bar()* method.

```python
import matplotlib.pyplot as plt
import numpy         

scores = [554, 654, 606]
diff_scores = [456, 589, 615]


index = numpy.arange(3)

plt.bar(index, diff_scores, width=.5,) #manipulating the width to make both bars visible
plt.bar(index, scores, width=0.2,color=(.5, .5, .7))
plt.grid(True)

plt.show()
```

This does not make the graph look as nice. This can also be changed by shifting the bar over. This can be done by adding or subtracting from numpy variable being passed into the *bar()* method.

```python
import matplotlib.pyplot as plt
import numpy         

scores = [554, 654, 606]
diff_scores = [456, 589, 615]


index = numpy.arange(3)

plt.bar(index + 0.2, diff_scores, width=.2) # adding the same value to index as bar width so bars can be side by side
plt.bar(index, scores, width=0.2,color=(.5, .5, .7)) 
plt.grid(True)

plt.show()
```

## Legend, Labels and Realigning axis

Printing the legend can be done by passing in the *labels* parameter into the *bar()* method and setting it equal to a string. Then, call the *legend()* method from the pyplot library and it will display a legend on top corner. The x-axis can also be realigned using the *xticks()* method from pyplot. In the parenthesises of *xticks()*, the index of the graph can be realigned by adding or subtracting from the index. Another parameter can be passed as a list which will label all the bars on the chart. On a side note, a cool parameter that is taken within the *bar()* is the *alpha*. *alpha* can be set to any value from 0-1.0 and it will determine the saturation of the bar in the chart.

```python
import matplotlib.pyplot as plt
import numpy         

us_scores = [554, 654, 606]
china_scores = [456, 589, 615]
korea_scores = [436, 607, 660]
uk_scores = [605, 607, 435]


index = numpy.arange(3)

plt.bar(index, us_scores, width=.2, alpha=.7, label="US")
plt.bar(index + 0.2, china_scores, width=0.2, alpha=.7, label="China")
plt.bar(index + 0.4, korea_scores, width=.2, alpha=.7, label="Korea") 
plt.bar(index + 0.6, uk_scores, width=0.2, alpha=.7, label="UK") 
plt.grid(True)
plt.legend()

plt.xlabel("Subject")
plt.ylabel("Scores")
plt.title("PISA Test Scores")

plt.xticks(index + .275, ["Mathematics", "Science", "English"])
plt.show()
```

Text can also be placed anywhere on the graph. This can be done using the *text()* method from pyplot. We are going to use this method to place each data value above the bar representing it. Here are some parameters that can be passed into *text()*. -- https://matplotlib.org/api/_as_gen/matplotlib.pyplot.text.html

```python
import matplotlib.pyplot as plt
import numpy         

us_scores = [554, 654, 606]
china_scores = [456, 589, 615]
korea_scores = [436, 607, 660]
uk_scores = [605, 607, 435]


index = numpy.arange(3)

# setting bar charts equal to variable so we can get data from it and print on bar chart
US = plt.bar(index, us_scores, width=.2, alpha=.7, label="US")
China = plt.bar(index + 0.2, china_scores, width=0.2, alpha=.7, label="China")
Korea = plt.bar(index + 0.4, korea_scores, width=.2, alpha=.7, label="Korea") 
UK = plt.bar(index + 0.6, uk_scores, width=0.2, alpha=.7, label="UK") 
plt.grid(True)
plt.legend()


plt.xlabel("Subject")
plt.ylabel("Scores")
plt.title("PISA Test Scores")

plt.xticks(index + .275, ["Mathematics", "Science", "English"])

def CreateNumLabels(data):    # creating function to print data value on bar chart
    for item in data:
        height = item.get_height()  # built in function from plt.bar(). get_x() and get_width() also can be used
        plt.text(item.get_x() + item.get_width() / 2, height, # position x ,and then y
                 height, ha="center", va="bottom") # ,printing height ,and then assigning vertical/horizontal alignment

CreateNumLabels(US)
CreateNumLabels(China)
CreateNumLabels(Korea)
CreateNumLabels(UK)

plt.show()
```

The legend in the chart can be manipulated many different ways. This can be done by passing in parameters. Here is the documentation for all parameters that can be passed into *pyplot.legend()*. https://matplotlib.org/3.3.1/api/_as_gen/matplotlib.pyplot.legend.html

Some of the *legend()* parameters I use in the code below which are also listed in the documentation are *frameon*,  *loc* and *bbox_to_anchor*.

```python
import matplotlib.pyplot as plt
import numpy         

us_scores = [554, 654, 606]
china_scores = [456, 589, 615]
korea_scores = [436, 607, 660]
uk_scores = [605, 607, 435]


index = numpy.arange(3)

US = plt.bar(index, us_scores, width=.2, alpha=.7, label="US")
China = plt.bar(index + 0.2, china_scores, width=0.2, alpha=.7, label="China")
Korea = plt.bar(index + 0.4, korea_scores, width=.2, alpha=.7, label="Korea") 
UK = plt.bar(index + 0.6, uk_scores, width=0.2, alpha=.7, label="UK") 
plt.grid(True)
plt.legend(frameon=True, loc=2, bbox_to_anchor=(1.0, 1.0))


plt.xlabel("Subject")
plt.ylabel("Scores")
plt.title("PISA Test Scores")

plt.xticks(index + .275, ["Mathematics", "Science", "English"])

def CreateNumLabels(data):    
    for item in data:
        height = item.get_height()  
        plt.text(item.get_x() + item.get_width() / 2, height, 
                 height, ha="center", va="bottom")
CreateNumLabels(US)
CreateNumLabels(China)
CreateNumLabels(Korea)
CreateNumLabels(UK)

plt.show()
```
